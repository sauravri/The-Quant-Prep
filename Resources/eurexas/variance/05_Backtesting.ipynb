## Data for Single Maturity

```python
import numpy as np
import pandas as pd
import datetime as dt
import warnings
warnings.simplefilter('ignore')

# @supress
import sys; sys.path.append('./scripts')

import index_date_functions as df

SUB_INDEXES = ['NA', 'V6I1', 'V6I2', 'V6I3',
             'V6I4', 'V6I5', 'V6I6', 'V6I7', 'V6I8']
```

```python
def get_riskless_rate():
    ''' Returns the riskless rate, in this case a constant value.
    '''
    return 0.02
```

```python
def load_vstoxx():
    ''' Loads all historical data of the VSTOXX and its sub-indexes from the 
    Internet and returns the data as a pandas DataFrame object.
    '''
    try:
        vstoxx_url = 'http://www.stoxx.com/download/historical_values/h_vstoxx.txt'
        vstoxx = pd.read_csv(vstoxx_url, index_col=0, parse_dates=True, 
                             dayfirst=True, header=2)
    except:
        # read stored data if there is no Internet connection
        vstoxx = pd.HDFStore('data/V2TX.h5', 'r')['V2TX']
    return vstoxx
```

```python
def load_stoxx():
    ''' Loads all historical data of the EURO STOXX from the Internet
    and returns the data as a pandas DataFrame object.
    '''
    try:
        stoxx_url = 'http://www.stoxx.com/download/historical_values/hbrbcpe.txt'
        col_name = ['Date', 'SX5P', 'SX5E', 'SXXP', 'SXXE', 'SXXF', 'SXXA', 
                    'DK5F', 'DKXF', 'DEL']
        stoxx = pd.read_csv(stoxx_url, index_col=0, parse_dates=True, 
                            dayfirst=True, sep=';', skiprows=4, 
                            names=col_name)
        stoxx = stoxx['SX5E']
    except:
        # read stored data if there is no Internet connection
        stoxx = pd.HDFStore('data/SX5E.h5', 'r')['SX5E']['SX5E']
    return stoxx
```

```python
stoxx = load_stoxx()
vstoxx = load_vstoxx()
```

```python
def get_subindex(running_month, running_year, maturity_month, maturity_year):
    ''' Returns the code of the VSTOXX sub-index to be used as implied volatility
    according to the time to maturity.
    
    Parameters
    ==========
    running_month: int
        the month of the current date
    running_year: int
        the year of the current date
    maturity_month: int
        the month of the maturity date
    maturity_year: int
        the year of the maturity date
        
    Returns
    =======
    sub_index_name: string
        name of the relevant sub-index
    '''
    # time to  maturity in months
    duration = maturity_month - running_month + (maturity_year - running_year) * 12
    
    # select the right sub-index code from the SUB_INDEX array defined in the header
    if duration < 1:
        sub_index_name = SUB_INDEXES[0]
    elif duration < 4:
        sub_index_name = SUB_INDEXES[duration]
    elif duration < 7:
        sub_index_name = SUB_INDEXES[4]
    elif duration < 10:
        sub_index_name = SUB_INDEXES[5]
    elif duration < 13:
        sub_index_name = SUB_INDEXES[6]
    elif duration < 19:
        sub_index_name = SUB_INDEXES[7]
    else:
        sub_index_name = SUB_INDEXES[8]
    return sub_index_name
```

```python
def make_future(start_month, start_year, maturity_month, maturity_year, stoxx, vstoxx):
    ''' Computes the time series data for the hypothectical settlement prices
    of a variance future defined by its starting date and maturity date.
    
    Parameter
    =========
    start_month: int
        the month of the start date.
    start_year: int
        the year of the start date.
    maturity_month: int
        the month of the maturity date.
    maturity_year: int
        the year of the maturity date.
    stoxx: pandas.DataFrame
        historical time series data of the EURO STOXX 50 index
    vstoxx: pandas.DataFrame
        historical time series data of the VSTOXX and its sub-indexes
    
    Returns
    =======
    data: pandas.DataFrame 
        object containing the hypothetical settlement prices
        of the variance futures contract
    '''
    # starting end settlement date
    start_date = df.third_friday(dt.datetime(start_year, start_month, 1))
    start_date = start_date - dt.timedelta(days=1)
    maturity = df.third_friday(dt.datetime(maturity_year, maturity_month, 1))
    end_date = min(maturity, dt.datetime.today()) 
    
    # the main dataframe
    data = pd.DataFrame(stoxx[(stoxx.index >= start_date)
                        & (stoxx.index <= end_date)])
    
    # observation days
    T = len(data)
    data['obs_days'] = np.arange(T)
    
    # implied volatility and variance
    data['sub_index'] = "NA"
    data['implied_vol'] = 0
    running_month = start_month
    running_year = start_year
    
    # the first sub-index for the implied volatility
    sub_index = get_subindex(running_month, running_year, maturity_month, maturity_year)
    data['implied_vol'].loc[start_date] = vstoxx[sub_index].loc[start_date]
    
    # runs through months to maturity
    while (running_month < maturity_month or 
           running_year < maturity_year):
        
        # select right sub-index
        sub_index = get_subindex(running_month, running_year, maturity_month, maturity_year)
        running_date = df.third_friday(dt.datetime(running_year, running_month, 1))
        
        # check if maturity lies in the future
        if running_date >= dt.datetime.today():
            break
            
        # when is a new sub-index used?
        running_end =  min(df.third_friday(running_date + dt.timedelta(days=20)),
                           dt.datetime.today())
        
        # apply the sub-index to that time window
        data['sub_index'][(data.index >= running_date) & (data.index < running_end)] = sub_index
        data['implied_vol'][(data.index >= running_date) & (data.index < running_end)] = vstoxx[sub_index] 
        
        # Go to the next month
        running_month += 1
        if running_month > 12:
            running_month -= 12
            running_year += 1
    
    # filling gaps 
    data = data.interpolate(method='time')
    if pd.isnull(data['implied_vol'].iloc[0]):
        delta = 1
        while pd.isnull(vstoxx[sub_index].loc[start_date-dt.timedelta(days=delta)]):
            delta += 1
            
        data['implied_vol'].iloc[0] = vstoxx[sub_index].loc[start_date-dt.timedelta(delta)]
        
    # squaring the volatility to get the variance
    data['implied_var'] = data['implied_vol'] ** 2
    
    # log returns of index
    data['returns'] = np.log(data['SX5E'] / data['SX5E'].shift(1))
    
    # realized variance
    data['realized_var'] = 252 * np.cumsum((data['returns'] * 100) ** 2) / data['obs_days']
        
    # strike
    strike = data['implied_vol'].iloc[0]
    
    # days to maturity
    data['days'] = [(maturity - day).days for day in data.index]
    
    # discount factors
    data['disc_factor'] = np.exp(- get_riskless_rate() * data['days'] / 365)
    
    # values
    data['pvs'] = (data['obs_days'] / T * data['realized_var']
                + (T - data['obs_days']) / T * data['implied_var'])
    data['value'] = (data['disc_factor'] * (data['obs_days'] / T * data['realized_var']
                + (T - data['obs_days']) / T * data['implied_var']  - strike ** 2) + 3000)
    
    # initial and final value
    data['value'].iloc[0] = 3000
    
    if end_date == maturity:
        # closing value for EURO STOXX 50 is used
        data['value'].iloc[-1] = data['realized_var'].iloc[-1] - strike ** 2 + 3000
    
    return data
```

## Data for All Maturities

```python
def generate_maturities(month, year):
    ''' Returns a list of all maturities of variance futures available at the release 
    date of a given month.
    
    Parameters
    ==========
    month: int
        the month of the observed date
    year: int
        the year of the observed date
    '''
    mat = set()

    # the subsequent 3 months
    for i in range(1, 4):
        add_month = month + i
        if add_month > 12:
            add_month -= 12
            add_year = year + 1
        else:
            add_year = year
        mat.add(add_year * 100 + add_month)
        
    # the subsequent 3 quarters
    for i in (3, 6, 9, 12):
        if month >= i:
            mat.add((year + 1) * 100 + i)
        else:
            mat.add(year * 100 + i)
    
    # and the subsequent half years
    for i in (6, 12):
        if month >= i:
            mat.add((year + 2) * 100 + i)
        else:
            mat.add((year + 1) * 100 + i)
    mat = list(mat)
    mat.sort()
    return mat
```

```python
def generate_all_series(start_month, start_year):
    ''' Creates time series for all variance futures starting on the 3rd Friday
    of a given month until now.
    
    Parameters
    ==========
    start_month: int
        the starting month
    start_year: int
        the starting year
    
    Returns
    =======
    all_data: pandas.DataFrame
        all time series data for all futures maturities; the dataframe has a double index, 
    the maturity month in the shape YYYYMM and the settlement date
    '''    
    # new DataFrame object
    all_data = pd.DataFrame()
    
    # the starting date
    issue_date = df.third_friday(dt.datetime(start_year, start_month, 1))
    # loop until today throiugh all months
    while issue_date < dt.datetime.today():
        
        # which maturities must exit?
        maturities = generate_maturities(start_month, start_year)
        for mat in maturities:
            
            # if the needed maturity does not exist we create it.
            if not mat in all_data.index:
                mat_year = int(mat / 100)
                mat_month = mat - mat_year * 100
                data = make_future(start_month, start_year, mat_month, mat_year, stoxx, vstoxx)
                
                # setting the settlement column
                data['Maturity'] = mat
                
                # drop the index of the dataframe
                data = data.reset_index()
                
                # create the new multi-index
                # data = data.set_index(['Maturity', 'Date'])
                
                # and append the new series to the main dataframe
                all_data = all_data.append(data)
            
        # compute the next observation day and continue the main loop
        start_month += 1
        if start_month > 12:
            start_month -= 12
            start_year += 1
        issue_date = df.third_friday(dt.datetime(start_year, start_month, 1))
    all_data = all_data.set_index(['Maturity', 'Date'])
    return all_data
```
