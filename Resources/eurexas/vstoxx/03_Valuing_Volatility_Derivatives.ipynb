## Preliminaries

```python
import sys
import numpy as np
import pandas as pd
sys.path.append('scripts')
path = './data/'
```

```python
# Model parameters
V0 = 17.5  # initial level of volatility index
kappa_V = 0.1  # speed of mean reversion
theta_V = 20.0  # long-term index level
sigma_V = 2.0  # volatility of volatility
zeta_V = 0.0  # factor of the expected volatility risk premium
r = 0.01  # risk-free interest rate

# Option parameters
K = 20.0  # strike
T = 1.0  # time horizon

# Simulation parameters
M = 50  # time steps
dt = T / M  # time interval
I = 50000  # number of MCS paths
```

## The Futures Pricing Formula

```python
import pricing_formulae as pf
```

```python
pf.futures_price(V0, kappa_V, theta_V, zeta_V, T)
```

## The Option Pricing Formula

```python
pf.call_price(V0, kappa_V, theta_V, sigma_V, zeta_V, T, r, K)
```

```python
option_values = []
strikes = np.linspace(15, 25)
for k in strikes:
    option_values.append(pf.call_price(V0, kappa_V, theta_V, sigma_V, zeta_V, T, r, k))
```

```python
data = pd.DataFrame(option_values, index=strikes, columns=['call values', ])
```

```python
%matplotlib inline
data.plot()
```

## Monte Carlo Simulation

```python
paths = pf.generate_paths(V0, kappa_V, theta_V, sigma_V, T, M, I)
```

```python
pd.DataFrame(paths[:, :10]).plot(legend=False)
```

```python
pf.call_estimator(V0, kappa_V, theta_V, sigma_V, T, r, K, M, I)
```

```python
%%time
estimates = []
for k in strikes:
    estimates.append(pf.call_estimator(V0, kappa_V, theta_V, sigma_V, T, r, k, M, I))
```

```python
data['estimates'] = estimates
```

```python
data.plot(style=['b', 'r.'])
```

```python
%%time
option_values = []
estimates = []
strikes = np.linspace(15, 25, 20)
for k in strikes:
    option_values.append(pf.call_price(V0, kappa_V, theta_V, sigma_V, 
                                       zeta_V, T, r, k))
    estimates.append(pf.call_estimator(V0, kappa_V, theta_V, sigma_V,
                                       T, r, k, M, I))
results = pd.DataFrame({'option_values': option_values,
                        'estimates': estimates}, index=strikes)

results.plot(style=['b', 'r.'])
```
